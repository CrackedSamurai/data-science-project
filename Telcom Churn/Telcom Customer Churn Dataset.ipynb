```python
highest_risk_customer = {}
high = data.groupby("gender").Churn.value_counts(normalize=True).loc[:, "Yes"] * 100
for col in data.columns:
    if col in ["customerID", "Churn", "tenure", "MonthlyCharges", "TotalCharges"]:
        continue
    else:
        try:
            high = data.groupby(col).Churn.value_counts(normalize=True).loc[:, "Yes"] * 100
            top_row = high.sort_values(ascending=False).head(1)
            highest_risk_customer[f"{col}"] = {"Category": top_row.index[0]
            , "Churn %" : round(top_row.values[0], 2)}
        except Exception as e:
            continue
highest_risk_customer = pd.DataFrame(highest_risk_customer).T.reset_index(names = ["Feature"])
print(highest_risk_customer)
```

```python
total_risk = []
for i in data.index:
    risk = []
    for col in data.columns:
        if col in ["customerID", "Churn", "tenure", "MonthlyCharges", "TotalCharges"]:
            continue
        else:
            try:
                high = data.groupby(col).Churn.value_counts(normalize=True).loc[:, "Yes"] * 100
                risk.append(high.loc[data.at[i, col]])
            except Exception as e:
                continue
    total_risk.append((round(np.mean(risk),2)))
data["risk_%"] = total_risk
```
